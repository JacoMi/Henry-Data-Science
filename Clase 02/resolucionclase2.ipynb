```python
import os
archivo = open('Emisiones_CO2.csv', 'r')

dicc_emisiones = {  'cod_pais' : [],
                    'nom_pais' : [],
                    'region' : [],
                    'anio' : [],
                    'co2' : [],
                    'co2_percapita' : []}
```

```python
next(archivo) #salta el header
for linea in archivo:

    linea = linea.strip() #borra último elemento de la linea

    #linea = linea.strip()

    campos = linea.split('|')

    dicc_emisiones['cod_pais'].append(campos[0])
    dicc_emisiones['nom_pais'].append(campos[1])
    dicc_emisiones['region'].append(campos[2])
    dicc_emisiones['anio'].append(campos[3])
    dicc_emisiones['co2'].append(campos[4])
    dicc_emisiones['co2_percapita'].append(campos[5])
```

```python
for indice, elemento in enumerate(dicc_emisiones['co2']): #indice para iterar durante la lista
    elemento= elemento.replace('.','')   # el enumerate crea un indice y el elemento en sí de la fila o lista
    elemento = elemento.replace(',','.') 
    if (elemento != ''):
        elemento= float(elemento)
    else:
        elemento = None
    dicc_emisiones['co2'][indice] = elemento

    #se utilizó para pasarlo a flotante los str de la anterior celda
```

```python
for indice, elemento in enumerate(dicc_emisiones['co2_percapita']):   #etl 
    elemento = elemento.replace('.','')  
    elemento = elemento.replace(',','.') 
    if (elemento != ''):
        elemento= float(elemento)
    else:
        elemento = None
    dicc_emisiones['co2_percapita'][indice] = elemento
```

```python
## 2D

nulo = 0 
for indice, elemento in enumerate(dicc_emisiones['cod_pais']):
    if elemento == None:
        nulo += 1
print('La cantidad de nulos para cod pais es ', nulo)

nulo = 0 
for indice, elemento in enumerate(dicc_emisiones['region']):
    if elemento == None:
        nulo += 1
print('La cantidad de nulos para region es ', nulo)

nulo = 0 
for indice, elemento in enumerate(dicc_emisiones['anio']):
    if elemento == None:
        nulo += 1
print('La cantidad de nulos para año es ', nulo)

nulo = 0 
for indice, elemento in enumerate(dicc_emisiones['co2']):
    if elemento == None:
        nulo += 1
print('La cantidad de nulos para co2 es ', nulo)

nulo = 0 
for indice, elemento in enumerate(dicc_emisiones['co2_percapita']):  #esté da 0 pero está mal REVISAR!!
    if elemento == None:
        nulo += 1
print('La cantidad de nulos para co2 per capita  es ', nulo)

#es el mismo código apra cada columna, panda lo tiene más optimizado pero etc 
```

```python
## 2 e

filtro_region = 'América Latina y Caribe'
filtro_anio = '2010'
emisiones = 0

for indice, elemento in enumerate(dicc_emisiones['region']):
    if (dicc_emisiones['region'][indice] == filtro_region):
        if (dicc_emisiones['anio'][indice] == filtro_anio):
            if (dicc_emisiones['co2'][indice] != None):
                emisiones += dicc_emisiones['co2'][indice]
print('Las emisiones en ' , filtro_region, ' en ', filtro_anio, ' fueron de ', emisiones, ' kt')
```
